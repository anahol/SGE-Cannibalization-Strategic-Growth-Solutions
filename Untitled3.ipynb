```python
# INSTALL LIBRARIES
!pip install pandas numpy matplotlib seaborn openpyxl -q

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# UPLOAD FILE (uncomment if not uploaded yet)
# from google.colab import files
# uploaded = files.upload()

print("="*80)
print("📊 CIAORENTCAR.COM — AI OVERVIEWS IMPACT ANALYSIS")
print("="*80)

# READ DATA
print("\n⏳ Loading data...")
df_timeline = pd.read_excel('Master_Dataset_Consolidated.xlsx', sheet_name='02_GSC_Timeline')
df_queries = pd.read_excel('Master_Dataset_Consolidated.xlsx', sheet_name='03_GSC_Queries')
df_pages = pd.read_excel('Master_Dataset_Consolidated.xlsx', sheet_name='04_GSC_Pages')
df_devices = pd.read_excel('Master_Dataset_Consolidated.xlsx', sheet_name='06_GSC_Devices')
df_bots = pd.read_excel('Master_Dataset_Consolidated.xlsx', sheet_name='10_User_Agents_Clean')
df_bookings = pd.read_excel('Master_Dataset_Consolidated.xlsx', sheet_name='09_Bookings')

print(f"✅ Loaded {len(df_timeline)} timeline rows")

# PARSE DATES & CTR
df_timeline['Date'] = pd.to_datetime(df_timeline['Date'])
df_timeline['CTR_numeric'] = df_timeline['CTR'].astype(str).str.rstrip('%').astype(float)
df_queries['CTR_numeric'] = df_queries['CTR'].astype(str).str.rstrip('%').astype(float)
df_pages['CTR_numeric'] = df_pages['CTR'].astype(str).str.rstrip('%').astype(float)

print("\n" + "="*80)
print("CZĘŚĆ 1: METRYKI — 3 PERSPEKTYWY")
print("="*80)

# PERSPECTIVE 1: March 2025 vs March 2026 (SGE Impact)
print("\n>>> PERSPEKTYWA 1: Marzec 2025 vs 2026 (SGE Launch Impact)")
mar_2025 = df_timeline[(df_timeline['Date'] >= '2025-03-01') & (df_timeline['Date'] <= '2025-03-31')]
mar_2026 = df_timeline[(df_timeline['Date'] >= '2026-03-01') & (df_timeline['Date'] <= '2026-03-31')]

def calc_metrics(df, label):
    clicks = df['Clicks'].sum()
    impressions = df['Impressions'].sum()
    ctr = df['CTR_numeric'].mean()
    position = df['Position'].mean()
    return {
        'Period': label,
        'Clicks': clicks,
        'Impressions': impressions,
        'CTR': ctr,
        'Position': position
    }

mar_metrics_2025 = calc_metrics(mar_2025, 'March 2025')
mar_metrics_2026 = calc_metrics(mar_2026, 'March 2026')

print(f"\n  Marzec 2025: {mar_metrics_2025['Clicks']:,} clicks, {mar_metrics_2025['CTR']:.2f}% CTR")
print(f"  Marzec 2026: {mar_metrics_2026['Clicks']:,} clicks, {mar_metrics_2026['CTR']:.2f}% CTR")

mar_clicks_pct = ((mar_metrics_2026['Clicks'] - mar_metrics_2025['Clicks']) / mar_metrics_2025['Clicks'] * 100)
mar_impr_pct = ((mar_metrics_2026['Impressions'] - mar_metrics_2025['Impressions']) / mar_metrics_2025['Impressions'] * 100)
mar_ctr_pct = ((mar_metrics_2026['CTR'] - mar_metrics_2025['CTR']) / mar_metrics_2025['CTR'] * 100)
mar_pos_pct = ((mar_metrics_2026['Position'] - mar_metrics_2025['Position']) / mar_metrics_2025['Position'] * 100)

print(f"  Zmiana: Clicks {mar_clicks_pct:+.1f}%, Impressions {mar_impr_pct:+.1f}%, CTR {mar_ctr_pct:+.1f}%")

# PERSPECTIVE 2: Q1 2025 vs Q1 2026 (Clean YoY)
print("\n>>> PERSPEKTYWA 2: Q1 2025 vs 2026 (Czysty rok do roku, bez kwietnia)")
q1_2025 = df_timeline[(df_timeline['Date'] >= '2025-01-01') & (df_timeline['Date'] <= '2025-03-31')]
q1_2026 = df_timeline[(df_timeline['Date'] >= '2026-01-01') & (df_timeline['Date'] <= '2026-03-31')]

q1_metrics_2025 = calc_metrics(q1_2025, 'Q1 2025')
q1_metrics_2026 = calc_metrics(q1_2026, 'Q1 2026')

print(f"\n  Q1 2025: {q1_metrics_2025['Clicks']:,} clicks, {q1_metrics_2025['CTR']:.2f}% CTR")
print(f"  Q1 2026: {q1_metrics_2026['Clicks']:,} clicks, {q1_metrics_2026['CTR']:.2f}% CTR")

q1_clicks_pct = ((q1_metrics_2026['Clicks'] - q1_metrics_2025['Clicks']) / q1_metrics_2025['Clicks'] * 100)
q1_impr_pct = ((q1_metrics_2026['Impressions'] - q1_metrics_2025['Impressions']) / q1_metrics_2025['Impressions'] * 100)
q1_ctr_pct = ((q1_metrics_2026['CTR'] - q1_metrics_2025['CTR']) / q1_metrics_2025['CTR'] * 100)

print(f"  Zmiana: Clicks {q1_clicks_pct:+.1f}%, Impressions {q1_impr_pct:+.1f}%, CTR {q1_ctr_pct:+.1f}%")

# PERSPECTIVE 3: Timeline (full year)
print("\n>>> PERSPEKTYWA 3: Timeline kompletny (Jan 2025 - Maj 2026)")
print(f"  Data początkowa: {df_timeline['Date'].min().strftime('%Y-%m-%d')}")
print(f"  Data końcowa: {df_timeline['Date'].max().strftime('%Y-%m-%d')}")
print(f"  Całkowite kliknięcia: {df_timeline['Clicks'].sum():,}")
print(f"  Kluczowe punkty: SGE launch (Mar 2025), New site (Apr 2026)")

# EXPORT ALL 3
summary_all = pd.DataFrame([
    {'Scenario': 'March 2025', **mar_metrics_2025},
    {'Scenario': 'March 2026', **mar_metrics_2026},
    {'Scenario': 'Q1 2025', **q1_metrics_2025},
    {'Scenario': 'Q1 2026', **q1_metrics_2026},
])
summary_all.to_csv('01_METRICS_3_PERSPECTIVES.csv', index=False)
print("\n✅ Exported: 01_METRICS_3_PERSPECTIVES.csv")

# ============================================================================
print("\n" + "="*80)
print("CZĘŚĆ 2: BOT ANALYSIS — Jakie boty crawlują stronę?")
print("="*80)

# Bot classification
def classify_bot(ua):
    ua_lower = str(ua).lower()
    if 'claude' in ua_lower:
        return 'ClaudeBot (Anthropic)'
    elif 'gpt' in ua_lower:
        return 'GPTBot (OpenAI)'
    elif 'google' in ua_lower:
        return 'GoogleBot'
    elif 'yandex' in ua_lower:
        return 'YandexBot'
    elif 'bingbot' in ua_lower:
        return 'BingBot'
    elif 'mj12bot' in ua_lower:
        return 'MJ12bot'
    elif 'applebot' in ua_lower:
        return 'AppleBot'
    elif 'bot' in ua_lower or 'crawl' in ua_lower:
        return 'Other Bots'
    else:
        return 'Real Users'

df_bots['Bot_Type'] = df_bots['User_Agent'].apply(classify_bot)

# Aggregate
bot_summary = df_bots.groupby('Bot_Type')['Occurrences'].sum().sort_values(ascending=False)
total_traffic = bot_summary.sum()

print(f"\nTraffic composition (monthly):")
print(f"{'Type':<30} {'Requests':>15} {'%':>10}")
print("-" * 55)
for bot_type, count in bot_summary.items():
    pct = count / total_traffic * 100
    print(f"{bot_type:<30} {count:>15,.0f} {pct:>9.1f}%")

# Bot vs User separation
ai_bots_count = bot_summary[bot_summary.index.str.contains('Claude|GPT|Google|Yandex|Bing|MJ12|Apple|Other', regex=True)].sum()
real_users = bot_summary.get('Real Users', 0)
bot_percentage = ai_bots_count / total_traffic * 100

print(f"\n🤖 BOT SUMMARY:")
print(f"   Total AI/Bot traffic: {ai_bots_count:,.0f} ({bot_percentage:.1f}%)")
print(f"   Real users: {real_users:,.0f} ({(real_users/total_traffic*100):.1f}%)")

# Export
bot_export = pd.DataFrame({
    'Bot_Type': bot_summary.index,
    'Occurrences': bot_summary.values,
    'Percentage': (bot_summary.values / total_traffic * 100).round(2)
})
bot_export.to_csv('02_BOT_ANALYSIS.csv', index=False)
print("\n✅ Exported: 02_BOT_ANALYSIS.csv")

# ============================================================================
print("\n" + "="*80)
print("CZĘŚĆ 3: QUERY ANALYSIS — Informational vs Transactional")
print("="*80)

# Classify queries
def classify_query(query):
    query_lower = str(query).lower()

    # Transactional keywords
    trans_keywords = ['ciao rent', 'prenota', 'book', 'reservation', 'noleggia', 'rent']

    # Informational keywords
    info_keywords = ['noleggio', 'scooter', 'auto', 'prezzo', 'costo', 'dove', 'come', 'cosa', 'affitto']

    for kw in trans_keywords:
        if kw in query_lower:
            return 'Transactional'

    for kw in info_keywords:
        if kw in query_lower:
            return 'Informational'

    return 'Mixed'

df_queries['Query_Type'] = df_queries['Query'].apply(classify_query)

# Summary by type
query_summary = df_queries.groupby('Query_Type').agg({
    'Clicks': 'sum',
    'Impressions': 'sum',
    'CTR_numeric': 'mean',
    'Position': 'mean'
}).round(2)

print(f"\nQuery type summary:")
print(query_summary)

# Top informational queries (zagrożone przez SGE)
info_queries = df_queries[df_queries['Query_Type'] == 'Informational'].nlargest(10, 'Clicks')
print(f"\n🔴 TOP INFORMATIONAL QUERIES (zagrożone SGE):")
print(info_queries[['Query', 'Clicks', 'Impressions', 'CTR_numeric', 'Position']].to_string())

# Top transactional (bezpieczne)
trans_queries = df_queries[df_queries['Query_Type'] == 'Transactional'].nlargest(10, 'Clicks')
print(f"\n🟢 TOP TRANSACTIONAL QUERIES (bezpieczne):")
print(trans_queries[['Query', 'Clicks', 'Impressions', 'CTR_numeric', 'Position']].to_string())

# Export
df_queries.to_csv('03_QUERY_ANALYSIS.csv', index=False)
print("\n✅ Exported: 03_QUERY_ANALYSIS.csv")

# ============================================================================
print("\n" + "="*80)
print("CZĘŚĆ 4: DEVICE ANALYSIS — Mobile vs Desktop Cannibalization")
print("="*80)

print(f"\nDevice performance:")
print(df_devices[['Device', 'Clicks', 'Impressions', 'CTR', 'Position']].to_string())

df_devices.to_csv('04_DEVICE_ANALYSIS.csv', index=False)
print("\n✅ Exported: 04_DEVICE_ANALYSIS.csv")

# ============================================================================
print("\n" + "="*80)
print("CZĘŚĆ 5: ROI CALCULATION — 3 Scenarios")
print("="*80)

# Scenarios data
scenarios = {
    'Scenario A: AIO SEO Only': {
        'cost_pr': 3000,
        'cost_it': 1500,
        'annual_revenue': 19375,
    },
    'Scenario B: AIO + Bot API': {
        'cost_pr': 3000,
        'cost_it': 3500,
        'annual_revenue': 24475,
    },
    'Scenario C: Full (AIO + Plugin)': {
        'cost_pr': 3000,
        'cost_it': 5000,
        'annual_revenue': 32500,  # estimate
    },
}

roi_results = []

for scenario_name, data in scenarios.items():
    total_cost = data['cost_pr'] + data['cost_it']
    year1_netto = data['annual_revenue'] - total_cost
    roi = (year1_netto / total_cost * 100)
    payback_months = total_cost / (data['annual_revenue'] / 12)

    roi_results.append({
        'Scenario': scenario_name,
        'Cost_Total': total_cost,
        'Annual_Revenue': data['annual_revenue'],
        'Year1_Netto': year1_netto,
        'ROI_%': roi,
        'Payback_Months': payback_months
    })

    print(f"\n{scenario_name}:")
    print(f"   Cost: €{total_cost:,}")
    print(f"   Year 1 Revenue: €{data['annual_revenue']:,}")
    print(f"   Year 1 Net: €{year1_netto:,}")
    print(f"   ROI: {roi:+.0f}%")
    print(f"   Payback: {payback_months:.1f} months")

roi_df = pd.DataFrame(roi_results)
roi_df.to_csv('05_ROI_SCENARIOS.csv', index=False)
print("\n✅ Exported: 05_ROI_SCENARIOS.csv")

# ============================================================================
print("\n" + "="*80)
print("CZĘŚĆ 6: VISUALIZATIONS")
print("="*80)

# Set style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (14, 10)

# Create figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(16, 10))

# Plot 1: Timeline - Clicks trend
ax1 = axes[0, 0]
timeline_monthly = df_timeline.groupby(df_timeline['Date'].dt.to_period('M'))['Clicks'].sum()
timeline_monthly.plot(ax=ax1, color='#1f77b4', linewidth=2, marker='o')
ax1.axvline('2026-04', color='red', linestyle='--', alpha=0.5, label='New website launch')
ax1.set_title('Clicks Trend (Jan 2025 - May 2026)', fontsize=12, fontweight='bold')
ax1.set_xlabel('Month')
ax1.set_ylabel('Clicks')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Plot 2: CTR trend
ax2 = axes[0, 1]
timeline_ctr = df_timeline.groupby(df_timeline['Date'].dt.to_period('M'))['CTR_numeric'].mean()
timeline_ctr.plot(ax=ax2, color='#ff7f0e', linewidth=2, marker='o')
ax2.axvline('2026-04', color='red', linestyle='--', alpha=0.5, label='New website launch')
ax2.set_title('CTR Trend (Jan 2025 - May 2026)', fontsize=12, fontweight='bold')
ax2.set_xlabel('Month')
ax2.set_ylabel('CTR (%)')
ax2.legend()
ax2.grid(True, alpha=0.3)

# Plot 3: Bot vs Users
ax3 = axes[1, 0]
bot_data = pd.Series({
    'AI/Bots': ai_bots_count,
    'Real Users': real_users
})
colors = ['#d62728', '#2ca02c']
ax3.pie(bot_data.values, labels=bot_data.index, autopct='%1.1f%%', colors=colors, startangle=90)
ax3.set_title('Traffic Composition: Bots vs Real Users', fontsize=12, fontweight='bold')

# Plot 4: ROI Comparison
ax4 = axes[1, 1]
scenarios_names = roi_df['Scenario'].str.replace('Scenario ', '')
roi_values = roi_df['ROI_%'].values
colors_roi = ['#1f77b4', '#ff7f0e', '#2ca02c']
bars = ax4.bar(scenarios_names, roi_values, color=colors_roi, alpha=0.7, edgecolor='black')
ax4.set_title('ROI Comparison by Scenario', fontsize=12, fontweight='bold')
ax4.set_ylabel('ROI (%)')
ax4.set_ylim(0, max(roi_values) * 1.2)
for i, (bar, val) in enumerate(zip(bars, roi_values)):
    ax4.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 20, f'{val:.0f}%',
             ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.savefig('06_ANALYSIS_CHARTS.png', dpi=300, bbox_inches='tight')
print("✅ Saved: 06_ANALYSIS_CHARTS.png")
plt.show()

# ============================================================================
print("\n" + "="*80)
print("✅ ANALYSIS COMPLETE!")
print("="*80)

print("\n📊 Exported files:")
print("  1️⃣  01_METRICS_YoY.csv")
print("  2️⃣  02_BOT_ANALYSIS.csv")
print("  3️⃣  03_QUERY_ANALYSIS.csv")
print("  4️⃣  04_DEVICE_ANALYSIS.csv")
print("  5️⃣  05_ROI_SCENARIOS.csv")
print("  6️⃣  06_ANALYSIS_CHARTS.png")

print("\n📥 Download all files to your computer")
print("   → Use them in Power BI and Notion")

print("\n" + "="*80)
```
